```python
vob = vb.ImageDescent(vb.surfaces.OUTWARD_GAUSS, (500,500),)

so.minimize(vob.step_noline, vob.params.flatten(), method='Nelder-Mead')

vob.plot()
```

```python
vob = vb.ImageDescent(vb.surfaces.OUTWARD_GAUSS, (500,500),)

optimizer = of.integrations.ScipyMinimize('nelder-mead', restart='last')

study = of.minimize(
    vob.step_noline,
    vob.params,
    optimizer,
    max_evals = 1000,
    bounds = (-1, 1)
)

vob.plot()
```
