```python
import optforge as of
from optforge.benchmark.vobench_benchmarks import VOBenchmark
import vobench as vb
import numpy as np

def rosenbrock(x, y):
    a = 1
    b = 100
    return (a - x) ** 2 + b * (y - x ** 2) ** 2
```

```python
vob = vb.FunctionDescent2D(rosenbrock, coords = [-1.9, -0.9], xrange = (-3, 3), yrange=(-2, 3))

optimizer = of.optim.RandomSearch()

study = of.minimize(
    vob.step_noline,
    vob.params,
    optimizer,
    max_evals = 1000,
    #bounds = (-1, 1)
)

vob.plot()
```
